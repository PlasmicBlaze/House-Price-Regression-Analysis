```r
# Select Variables for initial model and clean data
Train_Clean <- Train_data %>%
                    select_if(is.numeric)

Train_Clean <- Train_Clean %>%
               mutate_all(~replace_na(., mean(., na.rm = TRUE)))

Test_Clean <- Test_X %>%
                    select_if(is.numeric)

Test_Clean <- Test_Clean %>%
               mutate_all(~replace_na(., mean(., na.rm = TRUE)))

head(Train_Clean,5)
head(Test_Clean,5)

```
